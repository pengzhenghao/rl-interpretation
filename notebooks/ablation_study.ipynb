```python
from matplotlib import ticker
def draw(y_label, ax):
    baseline = result['default_policy/default_model/fc_out/no_ablation'][y_label]
    print("Baseline of measure {} is: {}. Min of data {}, Max of data {}".format(
        y_label, baseline, df[y_label].min(), df[y_label].max()))
    ax = sns.barplot(x="ablated_unit_index", y=y_label, color="k", data=df, ax=ax)
    ax.plot((0, 256), (baseline, baseline), color='r')
    ax.set_title(y_label)
    
    ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
    ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
#     ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

```python
sns.set_style("ticks")
plt.figure(figsize=(15, 15), dpi=300)

ax = plt.subplot(311)
y_label = "episode_reward_mean"
draw(y_label, ax)
plt.ylim(210, 310)

ax = plt.subplot(312)
y_label = "episode_length_mean"
draw(y_label, ax)
plt.ylim(800, None)

ax = plt.subplot(313)
y_label = "kl_divergence"
draw(y_label, ax)
plt.ylim(None, 400000)
plt.annotate("Max: 8267941.5", (75, 350000))
```

```python
with open("ablation_result_last2_layer_0830.pkl", 'rb') as f:
    result = pickle.load(f)
df = pd.DataFrame(result).T
```

```python
from matplotlib import ticker
def draw(y_label, ax):
    baseline = result['default_policy/default_model/fc2/no_ablation'][y_label]
    print("Baseline of measure {} is: {}. Min of data {}, Max of data {}".format(
        y_label, baseline, df[y_label].min(), df[y_label].max()))
    ax = sns.barplot(x="ablated_unit_index", y=y_label, color="k", data=df, ax=ax)
    ax.plot((0, 256), (baseline, baseline), color='r')
    ax.set_title(y_label)
    
    ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
    ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
#     ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

```python
sns.set_style("ticks")
plt.figure(figsize=(15, 15), dpi=300)

ax = plt.subplot(311)
y_label = "episode_reward_mean"
draw(y_label, ax)
plt.ylim(220, 310)

ax = plt.subplot(312)
y_label = "episode_length_mean"
draw(y_label, ax)
plt.ylim(800, None)

ax = plt.subplot(313)
y_label = "kl_divergence"
draw(y_label, ax)
# draw the outliners
ylim = 400000
plt.ylim(None, ylim)
ouliters = df[df.kl_divergence>1e6]
for loc, kl in zip(ouliters["ablated_unit_index"], ouliters["kl_divergence"]):
    plt.annotate(str(kl), (loc, 0.9* ylim))
```
