```python
agent_names = plot_df.agent.unique()
for name in agent_names:
    agent_pure_name = name.split(" rew")[0]
    iter_str = name.split(" ")[-1] # e.g. "iter=78"
    assert iter_str.startswith("iter=")
    num_iter = eval(iter_str.split("=")[-1])
    plot_df.loc[plot_df.agent==name, "iter"] = num_iter
    plot_df.loc[plot_df.agent==name, "name"] = agent_pure_name
plot_df.iter = plot_df.iter.astype(int)
pure_names = plot_df.name.unique()
for pure_name in pure_names:
    last_name = plot_df.loc[plot_df.name==pure_name].iloc[-1].agent
    plot_df.loc[plot_df.name==pure_name, "name"] = last_name

```
